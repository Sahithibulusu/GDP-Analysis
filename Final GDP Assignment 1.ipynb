```python
#Percentage Contribution of Primary, Secondary and teritiary Sectors as a percentage of total GDP of all states

df_total_GDP=df_all_states.loc[(df_all_states['Item']=='Gross State Domestic Product')][['2014-15','State']] #selecting required columns
```

```python
df_total_GDP.head()
```

```python
df_prim=df_all_states.loc[(df_all_states['Item']=='Primary')][['2014-15','State']].rename(columns={'2014-15':'Primary'}) #selecting primary and renaming the corresponding primary rows to primary
```

```python
df_sec=df_all_states.loc[(df_all_states['Item']=='Secondary')][['2014-15','State']].rename(columns={'2014-15':'Secondary'}) #selecting Secondary and renaming the corresponding Secondary rows to Secondary
```

```python
df_ter=df_all_states.loc[(df_all_states['Item']=='Tertiary')][['2014-15','State']].rename(columns={'2014-15':'Tertiary'}) #selecting Tertiary and renaming the corresponding Tertiary rows to Tertiary
```

```python
#Merging df_prim,df_sec,df_ter
df_prim_sec=pd.merge(df_prim,df_sec,how='inner',on='State')#merging primary and secondary
df_prim_sec_ter=pd.merge(df_prim_sec,df_ter,how='inner',on='State')#merging o/p of primary and secondary to teritiary
```

```python
df_prim_sec_ter.head()
```

```python
#merging the above o/p to total gdp
df_gdp_prim_sec_ter=pd.merge(df_total_GDP,df_prim_sec_ter,how='inner',on='State')
```

```python
df_gdp_prim_sec_ter.head()
```

```python
#percentage calculation of primary secondary and teritiary
df_gdp_prim_sec_ter['%primary_contribution']=(df_gdp_prim_sec_ter['Primary']/df_gdp_prim_sec_ter['2014-15'])*100#creating new column for Primary % calculation
df_gdp_prim_sec_ter['%secondary_contribution']=(df_gdp_prim_sec_ter['Secondary']/df_gdp_prim_sec_ter['2014-15'])*100#creating new column for Secondary % calculation
df_gdp_prim_sec_ter['%teritiary_contribution']=(df_gdp_prim_sec_ter['Tertiary']/df_gdp_prim_sec_ter['2014-15'])*100#creating new column forTertiary % calculation
```

```python
#creating new column for % calculation in all sectors
df_gdp_prim_sec_ter['%Total_pri_sec_ter']=df_gdp_prim_sec_ter['%primary_contribution']+df_gdp_prim_sec_ter['%secondary_contribution']+df_gdp_prim_sec_ter['%teritiary_contribution']
```

```python
df_gdp_prim_sec_ter=df_gdp_prim_sec_ter.sort_values(by='%Total_pri_sec_ter',ascending=False)#sorting values 
```

```python
df_gdp_prim_sec_ter.head()
```

```python
#plotting stack-bar for % contribution of primary, secondary and teritiary sectors

df_gdp_prim_sec_ter.plot(kind='bar',x='State',y=['%primary_contribution','%secondary_contribution','%teritiary_contribution'],stacked=True)
plt.show()
```
